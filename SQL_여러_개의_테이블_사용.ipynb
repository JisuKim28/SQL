```python
import os
import pandas as pd
from pandasql import sqldf
mysql = lambda q: sqldf(q, globals())

bike = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SQL/bike_dataset.csv')
orders = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SQL/orders.csv')
customers = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SQL/customers.csv')
employees = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SQL/employees.csv')
ds_2020 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SQL/DS_JobsSalaries_2020.csv')
ds_2021 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SQL/DS_JobsSalaries_2021.csv')
```

기본키
- 테이블은 각각 하나의 기본키를 가짐
- NULL값을 허용하지 않음

외래키
- JOIN으로 다른 테이블과 관계를 맺는 경우, 다른 테이블의 기본키를 참고하는 컬럼


```python
# orderId가 기본키가 됨
mysql("""

SELECT *
FROM orders
LIMIT 10

""")
```

```python
# customerId가 기본키, employeesId가 외래키가 됨
mysql("""

SELECT *
FROM customers
LIMIT 10

""")
```

```python
# employeesId가 기본키가 됨
mysql("""

SELECT *
FROM employees
LIMIT 10

""")
```

JOIN

한 개 이상의 테이블과 테이블을 서로 연결하는 기법

- INNER: 교집합
- LETF: 왼쪽 테이블 기준
- RIGHT: 오른쪽 테이블 기준

```python
# 2개의 테이블 합치기
mysql("""

SELECT *
FROM orders AS o
JOIN customers AS c
ON o.customerId = c.customerId

""")
```

```python
# 3개의 테이블 합치기

mysql("""

SELECT *
FROM orders AS o
JOIN customers AS c
ON o.customerId = c.customerId
JOIN employees AS e
ON c.employeesId = e.employeesId

 """)
```

```python
# LEFT 조인으로 합치기

mysql("""

SELECT *
FROM orders AS o
LEFT JOIN customers AS c
ON o.customerId = c.customerId
ORDER BY o.customerId

""")
```

UNION

- 테이블을 세로로 붙이는 형태
- 위 테이블과 아래 테이블 컬럼명이 달라도 UNION을 적용할 수 있음
- 위 테이블과 아래 테이블 컬럼수가 다르면 UNION을 사용할 수 없음
- 위 테이블과 아래 테이블 컬럼의 데이터 타입이 서로 다르면 UNION을 사용할 수 없음

사용법

- UNION ALL: 중복
- UNION: 중복 제거

```python
# 2020년 테이블과 2021년 테이블을 합치기
mysql("""

SELECT *
FROM ds_2020
UNION ALL
SELECT *
FROM ds_2021

""")
```

```python
# 2020년 테이블과 2021년 테이블을 합치고, work_year, experience_level, job_title만 추출
mysql("""

SELECT work_year, experience_level, job_title
FROM ds_2020
UNION ALL
SELECT work_year, experience_level, job_title
FROM ds_2021
LIMIT 100

""")
```

```python
# 2020년 테이블과 2021년 테이블을 중복을 제거하여 합치고, work_year, experience_level, job_title만 추출
mysql("""

SELECT work_year, experience_level, job_title
FROM ds_2020
UNION ALL
SELECT work_year, experience_level, job_title
FROM ds_2021
LIMIT 100

""")
```

Subquery

- JOIN과 달리 새로운 컬럼이 필요하지 않을 때 사용
- 서브 쿼리 실행 -> 메인 쿼리 실행
- 서브 쿼리 내에서는 ORDER BY 지원되지 않음

```python
# 가장 작은 주문금액 확인
mysql("""

SELECT MIN(price)
FROM orders

""")
```

```python
# 가장 작은 주문금액인 61,545원보다 작은 한도를 가진 고객 정보 확인
mysql("""

SELECT *
FROM customers
WHERE creditLimit < 61545
""")
```

단일행 서브쿼리

- 서브쿼리의 실행 결과가 항상 1건 이하인 서브쿼리
- 단일행 비교연산자(=, !=, >, < 등)와 함께 사용

```python
# 가장 작은 주문금액보다 작은 한도를 가진 고객 정보 확인
mysql("""

SELECT *
FROM customers
WHERE creditLimit < (SELECT MIN(price)
                     FROM orders)
""")
```

다중행 서브쿼리
- 서브쿼리 실행 결과가 여러 건인 서브 쿼리
- 다중행 비교연산자(IN, NOT IN 등)와 함께 사용

```python
# 미국에 사는 고객의 주문 번호
mysql("""

SELECT orderId
FROM orders
WHERE customerId IN (SELECT CustomerId
                     FROM customers
                     WHERE country = 'USA')

""")
```

FROM절 서브쿼리

- 데이터 추출 결과를 FROM 절에서 하나의 테이블처럼 사용하고 싶을 때
- FROM 절에 서브쿼리를 사용하면 특정 조건식을 갖는 데이터 추출결과를 테이블처럼 사용할 수 있음

```python
# 고객별 마지막 제품의 구매일자
mysql("""
SELECT customerId
      , MAX(orderDate) AS max_date
FROM orders
GROUP BY 1
""")
```

```python
# 고객별 마지막 제품의 구매일자가 현재일을 기준으로 며칠 지났는지 확인
mysql("""
SELECT *
    , CAST(julianday('now') - julianday(max_date) AS INT) AS date_diff
FROM (SELECT customerId
            , MAX(orderDate) AS max_date
      FROM orders
      GROUP BY 1) AS main
""")
```
